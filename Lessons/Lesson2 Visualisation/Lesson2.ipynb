## 2. Взаимодействия признаков
### 2.1 Количественные с количественными

```python
df.columns  # вывод колонок - признаков
```

```python
feat = [f for f in df.columns if 'charge' in f]
feat
```

```python
df[feat].hist();
```

```python
sns.pairplot(df[feat]);
# По диагонали - диаграмма рассеяния
```

```python
# Заготовка для раскрашивания диаграммы рассеяния в разные цвета в зависимости от того, ушёл клиент или нет
new_color = df['Churn'].map({False: 'blue', True: 'orange'})
```

```python
# Диаграмма рассеяния двух признаков (matplotlib) раскрашенная по оттоку ПРИМЕР!
plt.scatter(df['Total eve charge'], df['Total intl charge'], color=new_color, label='название картинки');
# Подписи x и y
plt.xlabel('Вечерние начисления');  # Работают латех формулы plt.xlabel(r'$a^2 + b^2$');
plt.ylabel('Международные начисления');
# Подпись легенды картинки
plt.legend();
```

```python
# Построение двух диаграмм рассеяния. На одной ушедшие
plt.scatter(df[df['Churn']]['Total eve charge'], df[df['Churn']]['Total intl charge'], color='orange',
            label='Churn (ушедшие)');
# ~ (тильда) даёт отрицание, то есть в данном случае не ушедшие, а лояльные
plt.scatter(df[~df['Churn']]['Total eve charge'], df[~df['Churn']]['Total intl charge'], color='blue',
            label='Loyal (лояльные)');
# Подписи x и y
plt.xlabel('Вечерние начисления');  # Работают латех формулы plt.xlabel(r'$a^2 + b^2$');
plt.ylabel('Международные начисления');
# Подписи легенды картинки
plt.legend();
# Подпись картинки
plt.title('Распределение начислений для лояльных/ушедших');
```
